```python
from matplotlib import pyplot as plt
```

```python
#We need subplot to plot 2 graph, else both graph will be merge in one.
```

```python
plt.subplot(221)

plt.bar([1,3,5,7],[3,5,6,2], label = "ABC", width=0.5) # Label and width are defined
plt.title("My Graph")
plt.xlabel("Time Period")
plt.ylabel("Value")


plt.subplot(222)
plt.scatter([2,4,6,8], [4,9,6,7], label= "Variable_A", color="b")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Graph")

plt.legend()
plt.show()
```
