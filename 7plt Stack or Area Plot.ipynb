```python
from matplotlib import pyplot as plt
```

```python
# Similar to the line graph to track the progress, best used when you are tracking multiple group over some x axis
```

```python
days = [1,2,3,4]

matplot = [2,1,2,1]
pandas = [1,2,1,3]
numpy = [1,3,2,1]

plt.title("Area Plot or Stack")
plt.xlabel("Days")
plt.ylabel("Hours")

plt.stackplot(days, matplot, pandas, numpy)
plt.show()
```

```python
plt.plot([],[], label="Matplot", linewidth = 5)
plt.plot([],[], label="Pandas", linewidth = 5)
plt.plot([],[], label="Numpy", linewidth = 5)
# the above code has been used to show the labels for the value, hence [],[] has been used for this
plt.legend()
plt.show()
```

```python
plt.stackplot(days, matplot, pandas, numpy, colors = ["m","c","r"])
plt.plot([],[], label="Matlpot", color="m", linewidth = 5)
plt.plot([],[], label="Pandas", color="c",  linewidth = 5)
plt.plot([],[], label="Numpy", color="r",  linewidth = 5)
plt.title("Area Plot or Stack")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.legend()
plt.show()
```
