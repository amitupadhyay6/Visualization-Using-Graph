```python
from matplotlib import pyplot as plt
```

```python
# Similar to stack plot, to show parts like % of something or parts
```

```python
slices = [35,15,25,15,10]
activities = ["Linear_Algebra","Calculus","Stats & Probability","Algo", "Others"]
cols = ["c","m","y","g", "r"]

plt.pie(slices,
       labels=activities,
       colors=cols,
       startangle=90, # on angle to start the graph
       shadow=True, # shadow of the graph, make it true if required
       explode=(0,0,0,0.5,0), # to pull out some data a bit, in this we have taken out the 2nd element, as 0,0.1,0,0 and 
                            # 0.x value, #define how much you need to pull
       autopct="%1.1f%%") # to overlay the % on the graph itself
plt.title("ML Requirement")
plt.show()
```
