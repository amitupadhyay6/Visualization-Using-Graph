```python
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
style.use("ggplot") ## need to check about this

y = [10,16,5]
x = [2,5,8]
y1 = [6,15,5]
x1 = [3,6,9]


plt.plot(x,y, "g", label = "HDFC", linewidth=5) # Define g as green color line, linewidth as width of line
plt.plot(x1,y1, "c", label = "ICICI", linewidth=5) # Define c as blue color line, linewidth as width of line

plt.legend() # this will print the legend define in label in above command
plt.title("My Graph")
plt.xlabel("Time")
plt.ylabel("Money")
plt.grid(True, color="r") # To color the grid line as k - as black


plt.show()
```
